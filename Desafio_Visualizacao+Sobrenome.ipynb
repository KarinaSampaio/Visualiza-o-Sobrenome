## Histograma com salários

```python
import pandas as pd
import matplotlib.pyplot as plt

# Carregar o dataset
file_path = 'HRDataset_v14.csv'
df = pd.read_csv(file_path)

# Criar faixas salariais para o gráfico de barras
salary_bins = [40000, 50000, 60000, 70000, 80000, 90000, 100000, 110000, 120000, 130000, 140000]
salary_labels = ['40k-50k', '50k-60k', '60k-70k', '70k-80k', '80k-90k', '90k-100k', '100k-110k', '110k-120k', '120k-130k', '130k-140k']

# Criar uma nova coluna para as faixas salariais
df['SalaryRange'] = pd.cut(df['Salary'], bins=salary_bins, labels=salary_labels, right=False)

# Contar a frequência de cada faixa salarial
salary_range_counts = df['SalaryRange'].value_counts().sort_index()

# Criar gráfico de barras horizontal
plt.figure(figsize=(12, 8))
salary_range_counts.plot(kind='barh', color='skyblue', edgecolor='black')
plt.xlabel('Frequência')
plt.ylabel('Faixa Salarial')
plt.title('Distribuição de Salários por Faixa Salarial')
plt.grid(axis='x')
plt.show()


```

## Gráfico de violino com salários discriminados por gênero

```python
import seaborn as sns

# Gráfico de violino com salários discriminados por gênero
plt.figure(figsize=(10, 6))
sns.violinplot(x='GenderID', y='Salary', data=df)
plt.xlabel('Gênero')
plt.ylabel('Salário')
plt.title('Distribuição de Salários por Gênero')
plt.xticks(ticks=[0, 1], labels=['Feminino', 'Masculino'])
plt.show()
```

## Série temporal correspondente à data de contratação

```python
# Converter a coluna de data de contratação para datetime
df['DateofHire'] = pd.to_datetime(df['DateofHire'])

# Traçar a série temporal da data de contratação
plt.figure(figsize=(10, 6))
df['DateofHire'].value_counts().sort_index().plot()
plt.xlabel('Data de Contratação')
plt.ylabel('Número de Contratações')
plt.title('Série Temporal de Contratação')
plt.show()
```
